# 3.2 Opếration sur les tableaux 

    a) Écrire une fonction diff_list qui calcule la dérivée d'une liste et une fonction diff_np qui fait la même opération mais sur des vecteurs Numpy en utilisant le slicing

```python
# Dérivée d'une liste 

v=[1,4,10,17,26,35,38,49,100]

def diff_list(v):
    n = len(v)
    d = [v[i+1] - v[i] for i in range(n-1)]
    return d
        
diff_list(v)
```

```python
# Dérivée d'un vecteur Numpy

def diff_np (l):
    t=np.array(l,dtype=float)
    n=len(t)
    c=np.zeros(n-1,dtype=float)
    for i in range (len(t)-2):
        a=t[i:i+2]
        b=t[i+1:i+3]
        c[i:i+2]=b-a
    return c
    
```

    b) Compairaison du temps d'éxecution des deux deux fonctions 

```python
a_list = [np.random.random() for _ in range(1000)]
a_np = np.random.random(1000)

%%timeit diff_list(a_list)
```

# 3.3 Représentations graphiques 

Question a)

```python
import numpy as np
import matplotlib.pyplot as plt
import math as m


x=np.linspace(0,2*m.pi,100)
y1=np.cos(x)
y2=np.cos(2*x)
y3=np.cos(3*x)
y4=np.sin(x)
y5=np.sin(2*x)
y6=np.sin(3*x)
plt.plot(x,y1,label="cos(x)")
plt.plot(x,y2,label="cos(2x)")
plt.plot(x,y3,label="cos(3x)")
plt.plot(x,y4,label="sin(x)")
plt.plot(x,y5,label="sin(2x)")
plt.plot(x,y6,label="sin(3x)")
plt.xticks(np.arange(0, 2.5*m.pi, step=m.pi/2),('0','π/2','π','2π/3','2π'))
plt.yticks(np.arange(-1,1.25,step=0.25))
plt.title('Fonctions trigonométriques')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
